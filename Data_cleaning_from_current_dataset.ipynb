```r
library(readr)
data <- read_csv("current_dataset.csv")
for (i in 1:dim(data)[1]) {
  for (j in 1:dim(data)[2]) {
    if (is.na(data[i,j])){
      data[i,j]=0
    }
  }
}

trygpa = as.data.frame(matrix(0,dim(data)[1],12))
colnames(trygpa)=c(colnames(data)[20:27],'Gap','Slope','Valid term','Average')
for (i in which(rowMeans(data[,20:27])!=0)) {
  n0=which(data[i,20:27]!=0)
  fn0=min(which(data[i,20:27]!=0))
  ln0=max(which(data[i,20:27]!=0))
  if (length(n0)!=(ln0-fn0+1)){
    trygpa[i,9]=1
  }
  if (length(n0)!=0){
    for (j in 1:length(n0)) {
      trygpa[i,j]=round(data[i,19+n0][j],2)
    }
  }
  trygpa[i,10]=round(as.numeric(lm(as.numeric(trygpa[i,1:sum((data)[i,20:27]>0)])~as.numeric(1:sum((data)[i,20:27]>0)))$coefficients[2]),2)
  trygpa[i,11]=sum((data)[i,20:27]>0)
  trygpa[i,12]=round(sum(data[i,20:27])/sum((data)[i,20:27]>0),2)
}
trygpa[,9]=as.factor(trygpa[,9])

ap=as.data.frame(matrix(0,dim(data)[1],1)) 
for (i in 1:dim(data)[1]) {
  ap[i,1]=round((floor((as.numeric(data[i,14])-190000)/100)+((as.numeric(data[i,14])-190000)/100-floor((as.numeric(data[i,14])-190000)/100))/1.2)-(floor((as.numeric(data[i,5])-190000)/100)+((as.numeric(data[i,5])-190000)/100-floor((as.numeric(data[i,5])-190000)/100))/1.2),1)
}
colnames(ap)='Academic Period (Years)'

data[,c(14,20:27)]=NULL
data=cbind(data,trygpa,ap)

data$`Academic Period Matriculated`=floor(data$`Academic Period Matriculated`/100)

for (i in 1:nrow(data)) {
  for (j in 1:ncol(data)) {
    if (is.na(data[i,j])==1){
      data[i,j]=0
    }
  }
}

data$`Grad Funding Indicator`[which(data$`Grad Funding Indicator`==unique(data$`Grad Funding Indicator`)[1])]=1
#0 for no fund and 1 for funded
```

```r
df = as.data.frame(data)
colnames(df)
dim(df)
```
